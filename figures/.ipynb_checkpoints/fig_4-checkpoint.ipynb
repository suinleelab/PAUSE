```python
from torch.utils.data import Dataset, DataLoader
class RNASeqData(Dataset):
    
    def __init__(self, X, c=None, y=None, transform=None):
        self.X = X
        self.y = y
        self.c = c
        self.transform = transform
        
    def __len__(self):
        return self.X.shape[0]
    
    def __getitem__(self, index):
        sample = self.X[index,:]
        
        if self.transform is not None:
            sample = self.transform(sample)
            
        if self.y is not None and self.c is not None:
            return sample, self.y[index], self.c[index]
        if self.y is None and self.c is not None:
            return sample, self.c[index]
        else:
            return sample
```

```python
outputs = datVAE.model(torch.tensor(data.X).cuda())
```

```python
recons = anndata.AnnData(
    pd.DataFrame(
        outputs.global_recon.detach().cpu().numpy(),
        index=data.obs_names,
        columns=data.var_names),
    obs=data.obs,
    varm=data.varm,
)

recons.obsm['codes'] = pd.DataFrame(
    outputs.z.detach().cpu().numpy(),
    index=data.obs_names,
    columns=datVAE.latent_space_names())
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
def extract_pathway_cols(df, pathway):
    mask = df.columns.str.startswith(pathway + '-')
    return df.loc[:, mask]
```

```python
stim_score = []
for pathway in datVAE.model.terms:
    pathway_design_matrix = extract_pathway_cols(recons.obsm['codes'],pathway)
    y = (recons.obs['condition'] == 'stimulated').values
    clf = LogisticRegression(random_state=0).fit(pathway_design_matrix, y)
    stim_score.append(clf.score(pathway_design_matrix, y))
```

```python
pathway_importance = pd.DataFrame(index=datVAE.model.terms)
pathway_importance['lr_score'] = stim_score
```
