```python
import numpy as np
import pandas as pd 
import anndata 
import sys
import matplotlib.pyplot as plt
import seaborn as sb
import mygene
import torch

from sklearn.manifold import TSNE
from sklearn.model_selection import train_test_split
from utils import load_annotations
from torch.utils.data import Dataset, DataLoader
import argparse
from sklearn.linear_model import LogisticRegression
import time 

from datasets import RNASeqData
from models import pmVAEModel
from sklearn.preprocessing import OneHotEncoder
```

```python
def get_sorted_means(ig, lr):
    ig['means'] = np.abs(ig).mean(1)
    lr['means'] = np.abs(lr).mean(1)
    
    sorted_ig = ig['means'].sort_values(ascending=False)
    sorted_lr = lr['means'].sort_values(ascending=False)
    
    return sorted_ig, sorted_lr
```

```python
paths = np.loadtxt('compare_figs/norman_paths.csv',delimiter=',', dtype='str')
```

```python
path_names = list(paths) + ['AUXILIARY']
```

```python
ig = pd.read_csv('/homes/gws/aspiro17/pace_genome_bio/new_for_revision/new_res/norman_granulocyte-apoptosis_ig.csv').set_axis(path_names)
lr = pd.read_csv('/homes/gws/aspiro17/pace_genome_bio/new_for_revision/new_res/norman_granulocyte-apoptosis_lr.csv').set_axis(path_names)

ig['means'] = np.abs(ig).mean(1)
lr['means'] = np.abs(lr).mean(1)

ig['rank']= ig['means'].rank(ascending=False)
lr['rank']= lr['means'].rank(ascending=False)
```

```python
sorted_ig = ig['means'].sort_values(ascending=False)
sorted_lr = lr['means'].sort_values(ascending=False)
```

```python
compare = pd.DataFrame()
compare['path'] = path_names
compare['ig_rank'] = ig['rank'].to_list()
compare['lr_rank'] = lr['rank'].to_list()
compare['ig_mean'] = ig['means'].to_list()
compare['lr_mean'] = lr['means'].to_list()
compare['ig_rank_swapped'] = 1/compare['ig_rank']
compare['lr_rank_swapped'] = 1/compare['lr_rank']
compare = compare.drop(len(paths)-1)
p_sizes = np.loadtxt('compare_figs/norman_p_sizes.csv',delimiter=',', dtype='int')
compare['p_sizes'] = p_sizes
compare
```

```python
keep = pd.DataFrame()
keep['path'] = path_names
top_ig = compare.loc[compare['ig_rank'] <= 20]
top_lr = compare.loc[compare['lr_rank'] <= 20]
keep = top_ig.combine_first(top_lr)
keep = keep.iloc[:-1,:] # get rid of aux
```

```python
# granulocyte settings

from matplotlib import patches
from matplotlib.lines import Line2D

yel = sb.color_palette(palette='colorblind')[8]
lb = sb.color_palette(palette='colorblind')[9]
dg = sb.color_palette(palette='colorblind')[2]

p_sizes = keep['p_sizes']

a,b = 20,200
dot_sizes = a + ((p_sizes - np.min(p_sizes))*(b-a))/(np.max(p_sizes)-np.min(p_sizes))

ig_above = .0004
lr_above = .02 

ig_min = 0 
lr_min = .59

x_offset = 0.005

ig_cutoff = float(keep[keep['ig_rank'] == 10]['ig_mean'])
lr_cutoff = float(keep[keep['lr_rank'] == 10]['lr_mean'])

fig = plt.figure(figsize=[7,7])
plt.fill_between(np.linspace(lr_cutoff,np.max(keep['lr_mean'])+ lr_above, 1000) , ig_min, ig_cutoff, color=yel, alpha=0.1,
                 interpolate=True)
plt.fill_between(np.linspace(lr_cutoff,np.max(keep['lr_mean'])+ lr_above, 1000) , ig_cutoff, np.max(keep['ig_mean'])+ ig_above, color=dg, alpha=0.1,
                 interpolate=True)
plt.fill_between(np.linspace(lr_min,lr_cutoff, 1000) , ig_cutoff, np.max(keep['ig_mean'])+ ig_above, color=lb, alpha=0.1,
                 interpolate=True)

plt.scatter(keep['lr_mean'], keep['ig_mean'], s=dot_sizes, c='k', linewidth=0, alpha=0.7)

plt.hlines(xmin=lr_min, xmax=np.max(keep['lr_mean'])+lr_above, y=ig_cutoff, linestyle='--', color='k', alpha=0.5)
plt.vlines(ymin=ig_min, ymax=np.max(keep['ig_mean'])+ig_above, x=lr_cutoff, linestyle='--',color='k', alpha=0.5)

pathway_list = keep['path']

texts = []
to_text = list(keep.index)

included = [
 'REACTOME_NEUTROPHIL_DEGRANULATION',
 'REACTOME_INNATE_IMMUNE_SYSTEM', 'REACTOME_CELL_CYCLE_MITOTIC',
 'REACTOME_DEVELOPMENTAL_BIOLOGY',
 'REACTOME_METABOLISM_OF_RNA',
 'REACTOME_CELL_CYCLE',
 'REACTOME_INFECTIOUS_DISEASE', 
'REACTOME_METABOLISM_OF_POLYAMINES', 'REACTOME_CELL_SURFACE_INTERACTIONS_AT_THE_VASCULAR_WALL', 'REACTOME_CYTOKINE_SIGNALING_IN_IMMUNE_SYSTEM']

for i in to_text:
    name = pathway_list[i].split('REACTOME_')[-1]

    x = keep['lr_mean'][i]
    y = keep['ig_mean'][i]

    
    if pathway_list[i] == 'REACTOME_CELL_SURFACE_INTERACTIONS_AT_THE_VASCULAR_WALL': 
        texts.append(plt.text(x=(x+x_offset), y=y, s='INTERACTIONS_AT_THE_VASCULAR_WALL', fontdict={'size':7.5}))
        plt.scatter(x,y, s=dot_sizes[i], c='r', linewidth=0)
        
    elif pathway_list[i] == 'REACTOME_DEVELOPMENTAL_BIOLOGY': 
        texts.append(plt.text(x=(x-10*x_offset), y=y+.00005, s=name, fontdict={'size':7.5}))
        plt.scatter(x,y, s=dot_sizes[i], c='r', linewidth=0)
     
    else:
        if pathway_list[i] in included: 
            texts.append(plt.text(x=(x+x_offset), y=y, s=name, fontdict={'size':7.5}))
            plt.scatter(x,y, s=dot_sizes[i], c='r', linewidth=0)

            
plt.xlabel(r'LR Score', fontsize=12)
plt.ylabel(r'Loss Attribution', fontsize=12)

plt.text(x= lr_min+ .01, y=np.max(keep['ig_mean']) + 0.0002, s='Unique to PAUSE', fontdict={'size':12, 'color':'black'})
plt.text(x=lr_cutoff+ .01, y=np.max(keep['ig_mean']) + 0.0002 , s='Common to both', fontdict={'size':12, 'color':'black'})
plt.text(x=lr_cutoff + .01, y=0.00005, s='Unique to LR score', fontdict={'size':12, 'color':'black'})

plt.title('K562 Granuloctye Top Pathways', fontsize=14)
plt.savefig('/homes/gws/aspiro17/pace_genome_bio/new_for_revision/new_figs/norman_granuloctye_no_cond.pdf')
```
