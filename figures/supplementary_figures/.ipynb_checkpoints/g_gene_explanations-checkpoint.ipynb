```python
import numpy as np
import pandas as pd 
import anndata 
import matplotlib.pyplot as plt
```

```python
aux_0 = pd.read_csv('kang_remove_g/aux_0.csv').sort_values('means',ascending=False)
aux_1 = pd.read_csv('kang_remove_g/aux_1.csv').sort_values('means',ascending=False)
aux_2 = pd.read_csv('kang_remove_g/aux_2.csv').sort_values('means',ascending=False)
aux_3 = pd.read_csv('kang_remove_g/aux_3.csv').sort_values('means',ascending=False)
```

```python
# plot 
def make_top_genes_plot(dataset, title, number_top):
    means = list(dataset['means'])[:number_top]
    stds = list(dataset['stds'])[:number_top]

    bars = np.arange(len(means))
    labels = list(dataset['index'])[:number_top]

    #fig, ax = plt.subplots()
    fig, ax = plt.subplots(figsize=(6,3))

    #ax.barh(bars, means, xerr=stds, align='center', alpha=0.5, ecolor='black', capsize=10)
    ax.barh(bars, means, xerr=stds, align='center', alpha=0.5, ecolor='black')

    ax.set_yticks(bars)
    ax.set_yticklabels(labels)
    
    plt.xlim([0, 2.8])

    ax.invert_yaxis()  # labels read top-to-bottom
    ax.set_xlabel('Mean Absolute Gene Attribution')
    #plt.tight_layout()

    #plt.savefig(str(dataset)+'.pdf', bbox_inches='tight')
    plt.savefig('gene_attribs/' + title +'.pdf', bbox_inches='tight')

    plt.show()
```

```python
make_top_genes_plot(aux_0, 'g_aux_0', 10)
```

```python
make_top_genes_plot(aux_1, 'g_aux_1', 10)
```

```python
make_top_genes_plot(aux_2, 'g_aux_2', 10)
```

```python
make_top_genes_plot(aux_3, 'g_aux_3', 10)
```
